#Sales data Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#Reading the dataset

```python
data=pd.read_csv('salesforcourse-4fe2kehu.csv')
data
```

```python
data.describe()
```

```python
data.info()
```

#Handling missing values

```python
data.isnull().sum()
```

```python
data=data.drop('Column1',axis=1)
```

```python
data=data.drop('Date',axis=1)
```

```python
data
```

```python
data=data.dropna(axis=0)
```

```python
data
```

```python
data.isnull().sum()
```

#Adding new columns

```python
data['Profit']=data['Revenue']-data['Cost']
data
```

```python
data['Profit Margin']=data['Profit']/data['Revenue']
data
```

#data Visulization

```python
data['Country'].value_counts()
```

#Revenue generated by month in particular year

```python
sns.barplot(x='Revenue',y='Month',data=data,hue='Year')
plt.title('Revenue by Month in particular year')
plt.show()
```

#Revenue generated in each country

```python
import matplotlib.pyplot as plt

# Assuming 'data' is a DataFrame with 'Country' and 'Revenue' columns
data.groupby('Country')['Revenue'].sum().plot(kind='pie', autopct='%1.1f%%')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.title('Revenue by Country')
plt.show()
```

```python
data['Quantity'].value_counts()
```

#Revenue generated in each Category

```python
data.groupby('Product Category')['Revenue'].sum().plot(kind='pie', autopct='%1.1f%%')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.title('Revenue by Category')
plt.show()
```

```python
sns.barplot(y=data['Sub Category'],x=data['Revenue'])
plt.title('Revenue by Sub Category')
plt.show()
```

```python
sns.scatterplot(x=data['Cost'],y=data['Revenue'],hue=data['Year'])
plt.title('Relation Between Cost and Revenue')
plt.show()
```

```python
sns.scatterplot(x=data['Unit Price'],y=data['Revenue'],hue=data['Year'])
plt.title('Relation between Unit price and Revenue')
plt.show()
```

```python
data.groupby('Quantity')['Revenue'].sum().plot(kind='pie', autopct='%1.1f%%')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.title('Revenue by Quantity')
plt.show()
```

```python
 #Group by sub category and calculate total quantity sold
category_sales = data.groupby('Sub Category')['Quantity'].sum().reset_index()
```

```python
import plotly.express as px
fig = px.bar(category_sales, y='Sub Category', x='Quantity', text_auto='.2s',
            title="Product Sales Quantity Based on Sub Categories")
fig.show()
```

```python
 #Group by sub category and calculate total quantity sold
category_profit = data.groupby('Sub Category')['Profit'].sum().reset_index()
```

```python
fig = px.bar(category_profit, y='Sub Category', x='Profit', text_auto='.2s',
            title="Product Sales Profit Based on Sub Categories")
fig.show()
```

```python

# Assuming 'data' is a DataFrame with 'Country' and 'Revenue' columns
data.groupby('Country')['Profit'].sum().plot(kind='pie', autopct='%1.1f%%')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.title('profit by Country')
plt.show()
```

```python
fig = px.sunburst(data, path=['Product Category','Sub Category'], values='Profit')
fig.show()
```

```python
import pandas as pd
import plotly.express as px

# Assuming you have a DataFrame named 'data' with columns 'Sub Category' and 'Cost'
# Group by 'Sub Category' and sum the 'Cost' for each category
funnel_data = data.groupby('State')['Profit'].sum().reset_index()

# Sort the data by 'Cost' in descending order to create the funnel shape
funnel_data = funnel_data.sort_values(by='Profit', ascending=False)

# Create the funnel chart using Plotly Express
fig = px.funnel(funnel_data, x='Profit', y='State', title='Funnel Chart for profit by State')

# Optional: Customize the appearance of the funnel chart
#fig.update_traces(textinfo='percent previous+percent initial', hoverinfo='x+y')

# Show the funnel chart
fig.show()
```

```python
import pandas as pd
import plotly.express as px

# Assuming you have a DataFrame named 'data' with columns 'Sub Category' and 'Cost'
# Group by 'Sub Category' and sum the 'Cost' for each category
funnel_data = data.groupby('Sub Category')['Profit'].sum().reset_index()

# Sort the data by 'Cost' in descending order to create the funnel shape
funnel_data = funnel_data.sort_values(by='Profit', ascending=False)

# Create the funnel chart using Plotly Express
fig = px.funnel(funnel_data, x='Profit', y='Sub Category', title='Funnel Chart for profit by Sub Category')

# Optional: Customize the appearance of the funnel chart
#fig.update_traces(textinfo='percent previous+percent initial', hoverinfo='x+y')

# Show the funnel chart
fig.show()
```
